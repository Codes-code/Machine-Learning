# Basic Steps of Using Scikit-Learn API
### 1 Choose a class of model
### 2 Choose model hyperparameters
### 3 Arrage data into features matrix and target array
### 4 Fit model to data
### 5 Apply trained model to new data

# simple lenier regression


```python
x=10*np.random.rand(100)
y=3*x+np.random.randn(100)
```

```python
plt.scatter(x,y)
```

# Step 1
we have to decide which model to use. we have decided to use the linear regression model in this case. 

```python
from sklearn.linear_model import LinearRegression 
```

# Step 2
we decide on which parameters to pass in the model

```python
model= LinearRegression(fit_intercept=True)
```

```python
model
```

# Step 3
we reshape the data to we better suited for ML

```python
X=x.reshape(-1,1)
```

```python
x.shape
```

```python
X.shape
```

#### we reshaped the x matrix from group of data to a line of data. we did not reshape y for reasons unknown rn.

# Step 4
we simply fit the model to the data. y is our target array. 

```python
model.fit(X,y)
```

```python
model.coef_
```

# Step 5

### lemme understand

## i understand, we predicted the y using the current data. the predict function used the present data to find the relation between x and y. (which i think can be expressed using model.coef_) and the two graphs i have printed below. 

```python
x_fit=np.linspace(-1,11)
```

```python
X_fit=x_fit.reshape(-1,1)
```

```python
y_fit=model.predict(X_fit)
```

Now, were gonna compare the two. our orignal data, and the x,y inwhich y is the predicted one, predicted using our own data. 

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize=(12,8))
plt.scatter(x,y)
plt.plot(x_fit, y_fit)
```
